### 1. Create a class called `Book` that has the attributes `title`, `author`, `pages`, `price`. It also has the method `get_price()` which returns `price`. Then create two instances of this class:

- `b1`:
`title` -> 'Practical Programming...'
`author` -> 'Gries, Campbell, Montojo'
`pages` -> 383
`price` -> 50

- `b2`:
`title` -> 'Building a Career in Data Science'
`author` -> 'Robinson, Nolis'
`pages` -> 322
`price` -> 40

````python
# Your code goes here
class Book:
    def __init__(book,title,author,pages,price):
      book.title = title
      book.author = author
      book.pages = pages
      book.price = price
      
    def get_price(book):
      return book.price

b1 = Book('Practical Programming','Gries, Campbell, Montojo','383','50')
print('The book author names are ' + b1.author)
print('The book price is ' + b1.get_price())

b2 = Book('Building a Career in Data Science','Robinson, Nolis','322','40')
print('The book names is ' + b2.title)
print('The book price is ' + b2.get_price())
````

### 2. Create a `Phone` class that has the attributes `model`, `id`, `name`, `brand` and `price`. Create the getters and setters for each attribute. As well as the following methods: `calculate_total` which takes the price and calculates it by the tax argument.

````python
# Your code goes here
class Phone:
    def __init__(self,model,id,name,brand,price):
        self.model = model
        self.id = id
        self.name = name
        self.brand = brand
        self.price = price
    
    def calculate_total(self):
        ont_tax= 1.13
        total_price = self.price * ont_tax
        return total_price
    
p1 = Phone(2024,2.0,'sonata','hyndai',35000)
print('car name is ' + p1.name)
print(p1.model)
print(p1.calculate_total())
print('total price of ' +p1.name + ' is ' + str(p1.calculate_total()))
````

    def __init__(self, model, id, name, brand, price):
      self._model = model
      self._id = id
      self._name = name
      self._brand = brand
      self._price = price
    
    def calculate_total(self, tax):
      return self._price * tax
